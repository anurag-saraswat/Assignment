
## Training

```python
def sigmoid(z):
    return (1.0)/(1+np.exp(-z))
    
def predict(X,weights):
    """X -> mx(n+1) matrix, w -->(nX1,) vector"""
    z = np.dot(X,weights)
    predictions = sigmoid(z)
    return predictions

def loss(X,Y,weights):
    """Binary Cross Entropy"""
    Y_ = predict(X,weights)
    cost = np.mean(-Y*np.log(Y_) - (1-Y)*np.log(1-Y_))
    return cost

def update(X,Y,weights,learning_rate):
    """Perform weight updates for 1 epoch"""
    Y_ = predict(X,weights)
    dw = np.dot(X.T,Y_ - Y)
    
    m = X.shape[0]
    weights = weights - learning_rate*dw/(float(m))
    return weights

def train(X,Y,learning_rate=0.5,maxEpochs=1000):
    # initialising empty list to store loss
    ls = []  
    #Modify the input to handle the bias term
    ones = np.ones((X.shape[0],1))
    X = np.hstack((ones,X))
    
    #Init Weights 0
    weights = np.zeros(X.shape[1]) #n+1 entires
    
    #Iterate over all epochs and make updates   
    for epoch in range(maxEpochs):
        weights = update(X,Y,weights,learning_rate)
        l = loss(X,Y,weights)
        ls.append(l)
    return weights,ls
```

## (a) vary the learning rate and show the best learning rate value when your run it for 50 epochs. 

Varying learning rate between from (0,1] and finding best learning rate(which give less average error)

```python
lr = 0.1
l_rate = []
t_loss = []
min_loss = float('inf')
min_lr = 0.1

while(lr<1):
  weights1 , loss1 = train(x,y_0_all,learning_rate=lr,maxEpochs=50)
  weights2 , loss2 = train(x,y_2_all,learning_rate=lr,maxEpochs=50)

  temp = (np.mean(np.array(loss1)) + np.mean(np.array(loss2)))/2
          
  if (temp<min_loss) :
    min_loss = temp
    min_lr = lr
  
  l_rate.append(lr)
  t_loss.append(temp)
  lr = lr + 0.1


plt.style.use("seaborn")
plt.xlabel("Learning Rate")
plt.ylabel("Error")
plt.title("Change in error with varying learning rate")
plt.plot(l_rate,t_loss,c='red')
plt.scatter(min_lr,min_loss,c='blue',label="Best Learning Rate")
plt.legend()
plt.show()


```

# Q2. Implement a 3-class backpropagation NNet on your own to classify iris data, i.e. from scratch. You should not be using any inbuilt function for this implementation (except reading the data). 
(a) vary the learning rate and show the best learning rate value when your run it for 50 epochs. 
(b) vary the number of epochs from 10 to 100 in a step of 10 and show the loss value curve (using the best learning rate obtained from (a))
(c) add L2 regularization - show the comparisons with and without this regularization and analyze your results. 


```python
# Importing Library   
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
# Loading iris data set
data = datasets.load_iris()
x = np.array(data.data)
y = np.array(data.target)
print(x.shape)
print(y.shape)
print("Target name: ",data.target_names)
```

```python
class NeuralNetwork:
    def __init__(self,input_size,layers,output_size):
        np.random.seed(0)
        model = {}
        model['W1'] = np.random.randn(input_size,layers[0]) 
        model['b1'] = np.zeros((1,layers[0]))
        
        model['W2'] = np.random.randn(layers[0],layers[1])
        model['b2'] = np.zeros((1,layers[1]))
        
        model['W3'] = np.random.randn(layers[1],output_size)
        model['b3'] = np.zeros((1,output_size))
        
        self.model = model
        
    def forward(self,x):

        W1,W2,W3 = self.model['W1'],self.model['W2'],self.model['W3']
        b1,b2,b3 = self.model['b1'],self.model['b2'],self.model['b3']
        
        z1 = np.dot(x,W1) + b1
        a1 = np.tanh(z1)
        
        z2 = np.dot(a1,W2) + b2 
        a2 = np.tanh(z2)
        
        z3 = np.dot(a2,W3) + b3
        y_ = softmax(z3)
        
        #Store the outputs of each layer for back-propagation
        self.activation_outputs = (a1,a2,y_)
        
        return y_
        
    def backward(self,x,y,learning_rate=0.001):
        
        W1,W2,W3 = self.model['W1'],self.model['W2'],self.model['W3']
        a1,a2,y_ = self.activation_outputs
        
        delta3 = y_ - y
        
        dw3 = np.dot(a2.T,delta3)  
        db3 = np.sum(delta3,axis=0)
        
        delta2 = (1-np.square(a2))*np.dot(delta3,W3.T)  
        dw2 = np.dot(a1.T,delta2)  
        db2 = np.sum(delta2,axis=0)
        
        delta1 = (1-np.square(a1))*np.dot(delta2,W2.T) 
        dw1 = np.dot(x.T,delta1)
        db1 = np.sum(delta1,axis=0)
        
        self.model['W1'] -= learning_rate*dw1
        self.model['W2'] -= learning_rate*dw2
        self.model['W3'] -= learning_rate*dw3
        
        self.model['b1'] -= learning_rate*db1
        self.model['b2'] -= learning_rate*db2
        self.model['b3'] -= learning_rate*db3
        
        
        
    def predict(self,x):
        y_out = self.forward(x)
        return np.argmax(y_out,axis=1)
    
    def summary(self):
        W1,W2,W3 = self.model['W1'],self.model['W2'],self.model['W3']
        a1,a2,y_ = self.activation_outputs
        
        print("W1 ",W1.shape)
        print("A1 ",a1.shape)
        
        print("W2 ",W2.shape)
        print("A2 ",a2.shape)
        
        print("W3 ",W3.shape)
        print("Y_ ",y_.shape)        

def loss(y_oht,p):
    l = -np.mean(y_oht*np.log(p))
    return l    

def one_hot(y,depth):
    m = y.shape[0]
    y_oht = np.zeros((m, depth))
    y_oht[np.arange(m), y] = 1
    return y_oht

def softmax(a):
    ea = np.exp(a)
    return ea/np.sum(ea,axis=1,keepdims=True)

```

```python
def train(X,Y,model,epochs,lr):
    """Creates and Model and Trains it!"""
    
    val_loss = []
    
    classes = len(np.unique(Y))
    Y_OHT = one_hot(Y,classes)
    
    for ix in range(epochs):
        Y_ = model.forward(X)
        l = loss(Y_OHT,Y_)
        model.backward(X,Y_OHT,learning_rate=lr)
        #print("Epoch %d Loss %.4f"%(ix,l),end='\n')
        val_loss.append(l)
    return val_loss
```

## (a) vary the learning rate and show the best learning rate value when your run it for 50 epochs.

```python
min_loss = float('inf')
min_learning = 0
lr_list = []
ls_list = []
lr = 0
while(lr<0.002):
  model = NeuralNetwork(input_size=4,layers=[6,6],output_size=3)
  temp = train(x,y,model,50,lr)
  ls = np.mean(np.array(temp))
  lr_list.append(lr)
  ls_list.append(ls)
  if(ls<min_loss):
    min_loss = ls
    min_learning = lr
  lr+=0.0001

plt.style.use("seaborn")
plt.xlabel("Learning Rate")
plt.ylabel("Error")
plt.title("Change in error with varying learning rate")
plt.plot(lr_list,ls_list,c='red')
plt.scatter(min_learning,min_loss,c='blue',label="Best Learning Rate")
plt.legend()
plt.savefig('2_1.png')
plt.show()

```

## (b) vary the number of epochs from 10 to 100 in a step of 10 and show the loss value curve (using the best learning rate obtained from (a))

```python
ep = 10
ep_list = []
ep_ls_list = []
lr = min_learning
while(ep<=100):
  model = NeuralNetwork(input_size=4,layers=[4,4],output_size=3)
  temp = train(x,y,model,ep,lr)
  ls = np.mean(np.array(temp))
  ep_list.append(ep)
  ep_ls_list.append(ls)
  ep+=10

plt.style.use("seaborn")
plt.xlabel("Epoch")
plt.ylabel("Error")
plt.title("Change in error with varying epoch")
plt.plot(ep_list,ep_ls_list,c='red')
plt.savefig('2_2.png')
plt.show()
```

```python
# Importing Library
import torch
import torchvision
import torch.nn.init
import torch.nn as nn
import torchvision.utils as utils
from torch.autograd import Variable
from torchvision import transforms, datasets
```

## Data Preparation

```python
# loading MNIST dataset

train = datasets.MNIST('', train=True, download=True,transform=transforms.Compose([transforms.ToTensor()]))
test = datasets.MNIST('', train=False, download=True,transform=transforms.Compose([transforms.ToTensor()]))
```

```python

# Visualising 10 random sample from training data
indexes = np.random.randint(0, train.data.size()[0], size=10)

# plot the 10 mnist digits
plt.figure(figsize=(6, 6))
for i in range(len(indexes)):
    sub = plt.subplot(2, 5, i + 1)
    image = train.data[indexes[i]]
    plt.imshow(image, cmap='gray')
    plt.axis('off')
    sub.set_title("Label :" + str(int(train.targets[indexes[i]])))
plt.show()
plt.close('all')
```

```python
#Loading Dataset

trainset = torch.utils.data.DataLoader(train, batch_size=100, shuffle=True)
batch_images, batch_labels = next(iter(trainset))

testset = torch.utils.data.DataLoader(test, batch_size=100, shuffle=False)
batch_images_test, batch_labels_test = next(iter(testset))
```

```python
# Defining Model

class RbfNet(nn.Module):
    def __init__(self, centers, num_class=10):
        super(RbfNet, self).__init__()
        self.centers = centers
        self.num_centers = centers.size(0)
        self.num_class = num_class
    
        self.linear = torch.nn.Linear(self.num_centers, self.num_class, bias=True)
        self.beta = nn.Parameter(torch.ones(1,self.num_centers)/10)

    def radial_fun(self, batches):
        n_input = batches.size(0) # number of inputs
        A = self.centers.view(self.num_centers,-1).repeat(n_input,1,1)
        B = batches.view(n_input,-1).unsqueeze(1).repeat(1,self.num_centers,1)
        C = torch.exp(-self.beta.mul((A-B).pow(2).sum(2,keepdim=False).sqrt()))
        return C
    
    def forward(self, batches):
        radial_val = self.radial_fun(batches)
        class_score = self.linear(radial_val)
        return class_score

#input batch size
centers = batch_images
model = RbfNet(Variable(centers), num_class=10)
print(model)
```

```python
# Hyper-parameters
learning_rate = 0.01 
training_epochs = 50
batch_size=100

# Adam Optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
cost_func = torch.nn.CrossEntropyLoss()

loss_list = []
lr_list = []

while(learning_rate<0.5):
  temp_loss = []
  for epoch in range(training_epochs):
    avg_loss = 0
    total_batch = len(train) // batch_size

    for i, (batch_images, batch_labels) in enumerate(trainset):
        
        
        X = Variable(batch_images.view(-1, 28 * 28))
        Y = Variable(batch_labels)       # label is not one-hot encoded

        optimizer.zero_grad()             # Zero Gradient Container
        Y_prediction = model(X)           # Forward Propagation
        cost = cost_func(Y_prediction, Y) # compute loss
        cost.backward()                   # compute gradient
        optimizer.step()                  # gradient update

        avg_loss += cost / total_batch
        temp_loss.append(avg_loss.item())

  loss_list.append(np.mean(np.array(temp_loss)))
  lr_list.append(learning_rate)
  learning_rate += 0.01
```

```python
plt.style.use("seaborn")
plt.xlabel("Learning Rate")
plt.ylabel("Error")
plt.title("Change in error with varying learning rate")
plt.plot(lr_list,loss_list,c='red')
plt.show()
```

```python
# Test the Model
correct = 0
total = int(test.data.size()[0])
for images, labels in testset:
    images = Variable(images.view(-1, 28*28))
    outputs = model(images)
    _, predicted = torch.max(outputs.data, 1)
    correct += (predicted == labels).sum()

print("Accuracy is : ",(correct/total)*100)
      
    
```

#Q4. Using MNIST database, code Autoencoder model with three encoding and three decoding layers. Show the visualization of the feature maps. On the features, add a classifier to perform 10-class classification and show the training loss curve and test accuracy.

```python
import torch
import torchvision 
import torchvision.transforms as transforms
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable
from torchvision.utils import save_image
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import matplotlib.pyplot as plt
import numpy as np
import torchvision.utils as utils
from torch.autograd import Variable
from torchvision import transforms, datasets
```

```python
# Loading Training Data 
train = datasets.MNIST('', train=True, download=True,transform=transforms.Compose([transforms.ToTensor()]))
test = datasets.MNIST('', train=False, download=True,transform=transforms.Compose([transforms.ToTensor()]))
trainset = torch.utils.data.DataLoader(train, batch_size=100, shuffle=True)
batch_images, batch_labels = next(iter(trainset))

testset = torch.utils.data.DataLoader(test, batch_size=100, shuffle=False)
batch_images_test, batch_labels_test = next(iter(testset))
```

```python
# Defining Model
class AE(nn.Module):
    def __init__(self, **kwargs):
        super().__init__()
        self.encoder_hidden_layer = nn.Linear(
            in_features=kwargs["input_shape"], out_features=128
        )
        self.encoder_output_layer = nn.Linear(
            in_features=128, out_features=128
        )
        self.decoder_hidden_layer = nn.Linear(
            in_features=128, out_features=128
        )
        self.decoder_output_layer = nn.Linear(
            in_features=128, out_features=kwargs["input_shape"]
        )

    def forward(self, features):
        activation = self.encoder_hidden_layer(features)
        activation = torch.relu(activation)
        code = self.encoder_output_layer(activation)
        code = torch.sigmoid(code)
        activation = self.decoder_hidden_layer(code)
        activation = torch.relu(activation)
        activation = self.decoder_output_layer(activation)
        reconstructed = torch.sigmoid(activation)
        return reconstructed
```

```python
batch_size = 512
epochs = 50
learning_rate = 0.001

#Configuring Device
device = torch.device("cpu")

# create a model from `AE` autoencoder class
# load it to the specified device
model = AE(input_shape=784).to(device)

# create an optimizer object
# Adam optimizer with learning rate 1e-3
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# mean-squared error loss
criterion = nn.MSELoss()
```

```python
ls_list =[]
ep_list = []

for epoch in range(epochs):
    loss = 0
    for batch_features, _ in trainset:
        # reshape mini-batch data to [N, 784] matrix
        # load it to the active device
        batch_features = batch_features.view(-1, 784).to(device)
        
        # reset the gradients back to zero
        # PyTorch accumulates gradients on subsequent backward passes
        optimizer.zero_grad()
        
        # compute reconstructions
        outputs = model(batch_features)
        
        # compute training reconstruction loss
        train_loss = criterion(outputs, batch_features)
        
        # compute accumulated gradients
        train_loss.backward()
        
        # perform parameter update based on current gradients
        optimizer.step()
        
        # add the mini-batch training loss to epoch loss
        loss += train_loss.item()
    
    # compute the epoch training loss
    loss = loss / len(trainset)
    
    ls_list.append(loss)
    ep_list.append(epoch)
    # display the epoch training loss
    print("epoch : {}/{}, recon loss = {:.8f}".format(epoch + 1, epochs, loss))
```

```python
plt.style.use("seaborn")
plt.xlabel("Learning Rate")
plt.ylabel("Error")
plt.title("Change in error with Epoch")
plt.plot(ep_list,ls_list,c='red')
plt.savefig('4_1.png')
plt.show()
```
